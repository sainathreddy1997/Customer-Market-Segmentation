### Preprocessing the data 

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled.shape
```

# Unsupervised Machine Learning
## K-Means Clustering

### Applying the Eblow method to find the optimal number of clusters
#### By plotting the # of clusters on x-axis vs within cluster sum of the squares on y-axis to find the optimal number of clusters that fit the data

```python
x,scores = [],[]
for i in range(1,15):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_scaled)
    scores.append(kmeans.inertia_)
    x.append(i)
plt.plot(x,scores,'bx-')
```

```python
# From the above plot we can notice that the optimal number of clusters is around 7
```

```python
# Building a KMeans clustering with 7 clusters
```

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(df_scaled)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = kmeans.cluster_centers_
```

```python
labels
```

```python
cluster_centers
```

```python
inv_cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(inv_cluster_centers,columns = [df.columns])
cluster_centers
```

#### From the above cluster centroids we can classify the customers into four main categories Transactors, New customers, Prime Customers and Revolvers.
##### Transactors - Customers who use their money in a careful manner and their balance will be low. Balance with nearly 110 dollars comes under this category 
##### Prime Customers - Customers whose balance will be high as well as purchasing frequency will be high. Balance nearly 5400 dollars and purchasing frequency of around 90% comes under this category
##### New Customers - Customers who are new compared to the other customers, genearally their tenure be low. 
##### Revolvers - Customers who uses the credit card as a loan. Generally their balance will be high and purchasing frequency will be low. Balance around 5000 dollars and purchasing frequency of 30% falls under this category

```python
y_predict = kmeans.predict(df_scaled)
```

```python
y_predict
```

#### Concatening the original dataframe with the clustered labels which are generated by applying KMeans algorithm

```python
df = pd.concat([df, pd.DataFrame(y_predict,columns = ['label'])], axis = 1)
```

```python
df
```

#### Plotting histograms for different clusters

```python
for i in df.columns:
    plt.figure(figsize=(40,5))
    for j in range(7):
        plt.subplot(1,7,j+1)
        cluster = df[df['label']==j]
        cluster[i].hist(bins = 25)
        plt.title('{} \n Cluster {}'.format(i,j))
    plt.show()
```

### Applying Dimensionality reduction using PCA

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)
principal_components
```

#### Creating a dataframe that consists pricipal components features and clustered labels

```python
pca_df = pd.DataFrame(data = principal_components, columns = ['pca1','pca2'])
```

```python
pca_df = pd.concat([pca_df, pd.DataFrame({'label':y_predict})], axis = 1)
```
